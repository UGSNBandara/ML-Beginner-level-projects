## Diabetes Risk Variation with Age

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots(figsize = (8,5))

ax.hist(diebetics_data["Age"], bins = 15)

ax.set(title="Diabetes Risk Variation with Age", xlabel="Age", ylabel="Frequent")
```

## Diabets Risk vs Gender 

```python
gender_risk = pd.DataFrame(diebetics_data["Gender"].value_counts())

fig, ax = plt.subplots(figsize = (6,5))

ax.bar(["Female", "Male"],gender_risk["count"],width=0.68)

ax.set(title="Diabets Risk vs Gender", xlabel="Gender", ylabel="Number of Diebtics")

```

```python
fig.savefig("Saved_Graphs/Diebets Risk vs Gender.png")
```

## Analyzing Symptom Frequency in Diabetes 

```python
symptom_df = diebetics_data.drop(["Gender","Age","class",], axis = 1)
## symptom_df --> Only consist the symptoms 
symptom_df
```

```python
freq = np.array(symptom_df["Polyuria"].value_counts()["Yes"])

for i in symptom_df.drop("Polyuria", axis=1).columns:
    freq = np.append(freq,symptom_df[i].value_counts()["Yes"])

freq
```

```python
fig, ax = plt.subplots(figsize = (17,8))

ax.bar(symptom_df.columns, freq, width=0.5)

ax.set_xlabel("Symptoms", fontsize = 18)
ax.set_ylabel("Frequency of Occurence", fontsize = 18)
ax.set_title("Frequency of Symptom Occurrence in Diabetes Patients", fontsize = 18)

description_text = (
    "This bar chart illustrates the frequency of symptoms observed in the group of diabetes patients. "
)
plt.figtext(0.5, -0.05, description_text, wrap=True, horizontalalignment='center', fontsize=14)
```

```python
fig.savefig("Saved_Graphs/symptoms_freq_of_diabetes_parients.png")
```

```python
x = diabetes_data.drop("class", axis=1)
y = diabetes_data["class"]

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


one_hot = OneHotEncoder()
x_noage = x.drop("Age", axis=1)
catogorical_columns = x_noage.columns


transformer = ColumnTransformer([("one_hot", one_hot, catogorical_columns)], remainder="passthrough")
transformed_x = transformer.fit_transform(x)
transformed_x_df = pd.DataFrame(transformed_x)


```

```python
np.random.seed(seed = 42)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(transformed_x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()

model1.fit(x_train, y_train)
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

### 0.8923 is a good level of score for a model
