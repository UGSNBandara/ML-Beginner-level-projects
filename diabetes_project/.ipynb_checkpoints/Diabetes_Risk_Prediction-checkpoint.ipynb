## Analyzing Symptom Frequency in Diabetes 

```python
symptom_df = diebetics_data.drop(["Gender","Age","class",], axis = 1)
## symptom_df --> Only consist the symptoms 
symptom_df
```

```python
freq = np.array(symptom_df["Polyuria"].value_counts()["Yes"])

for i in symptom_df.drop("Polyuria", axis=1).columns:
    freq = np.append(freq,symptom_df[i].value_counts()["Yes"])

freq
```

```python
fig, ax = plt.subplots(figsize = (17,8))

ax.bar(symptom_df.columns, freq, width=0.5)

ax.set_xlabel("Symptoms", fontsize = 18)
ax.set_ylabel("Frequency of Occurence", fontsize = 18)
ax.set_title("Frequency of Symptom Occurrence in Diabetes Patients", fontsize = 18)

description_text = (
    "This bar chart illustrates the frequency of symptoms observed in the group of diabetes patients. "
)
plt.figtext(0.5, -0.05, description_text, wrap=True, horizontalalignment='center', fontsize=14)
```

## Lets train a model to predict : 

```python
diabetes_data
```

```python
x = diabetes_data.drop("class", axis=1)
y = diabetes_data["class"]

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


one_hot = OneHotEncoder()
x_noage = x.drop("Age", axis=1)
catogorical_columns = x_noage.columns


transformer = ColumnTransformer([("one_hot", one_hot, catogorical_columns)], remainder="passthrough")
transformed_x = transformer.fit_transform(x)
transformed_x_df = pd.DataFrame(transformed_x)


```

```python
y_l = []
for ys in y:
    if ys == "Positive":
        y_l.append(1)
    elif ys == "Negative":
        y_l.append(0)
y_numeric = pd.Series(y_l)
y_numeric
```

```python
np.random.seed(seed = 42)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(transformed_x,y_numeric)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()

model1.fit(x_train, y_train)
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

### 0.8923 is a good level of score for a model

```python
## Lets save this model
import pickle
pickle.dump(model1, open("Saved_models/KNeighbor_model.pkl", "wb"))
```

#### Evaluation : 

Cross validation test:

```python
from sklearn.model_selection import cross_val_score
np.random.seed(42)

cross_score = np.mean(cross_val_score(model1, transformed_x, y, cv=5))
cross_score
```

```python
model1.predict(x_test)
```

```python
class_prob = model1.predict_proba(x_test)
positive_prob = class_prob[:,1]
```

```python
positive_prob
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, positive_prob)
```

```python
fpr.size
```

```python
## reason for 7 size: because most of tha cases are unique and only 7 thresholds 
## unique properbility scores 
```

### ROC curve

```python
fig, ax = plt.subplots(figsize=(6,5))

ax.plot(fpr, tpr , color = "red", label="ROC")
ax.plot([0,1], [0,1] , color = "darkblue", linestyle="--", label="Baseline-no predicting power")

ax.set_xlabel("False positive rate - fpr")
ax.set_ylabel("True positive rate - tpr")
ax.legend()
ax.set_title("Receiver Operating Characteristics (ROC) - model1")
```

```python
fig.savefig("Saved_Graphs/roc_curve_model1.png")
```

### Confusion matrix

```python
class_pred = model1.predict(x_test)
from sklearn.metrics import confusion_matrix
con_matrix = confusion_matrix(y_test, class_pred)
con_matrix
```

```python
##cross tab
pd.crosstab(y_test, class_pred, colnames=["Predicted Values"], rownames=["True Values"])
```

```python
#ploting the confusion metrix
import seaborn as sns
fig, ax = plt.subplots(figsize=(6,6))
ax  = sns.heatmap(con_matrix, annot=True, cbar=False, cmap='Reds')
ax.set(xlabel="Predicted Values", ylabel="True Valuse", title="Confusion Matrix - model1")
```

```python
fig.savefig("Saved_Graphs/confusion_matrix_model1.png")
```
